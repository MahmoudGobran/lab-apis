**Hint Section for 3. **Discover Related Artists**:**

1. **Getting Artist ID**:
    - Remember that every artist on Spotify has a unique identifier: their `id`. To get the related artists, you first need to fetch the ID of the given artist.
    - Consider using the `sp.search` method to query the artist's name. The method requires a `q` parameter, which is your query (in this case, the artist's name). It also has a `limit` parameter, which specifies the number of tracks it returns. In this case, 1 track is enough, since we just want the artist ID.

3. **Fetching Related Artists**:
    - Once you have the artist's ID, you can use another SpotiPy method to fetch related artists. Think about which SpotiPy method allows you to get related artists using an artist's ID. Here is the documentation link: https://spotipy.readthedocs.io/en/2.22.1/. 
    - This method will return a list of related artists. You'll need to extract the relevant details (artist names) from this list.

    - Always test your function with one artist name first. Once you're confident it works, then apply it to the entire list.

Remember, the key is to break the problem down into manageable steps. Use the SpotiPy documentation as a resource to understand available methods and their return structures.

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from requests.exceptions import ReadTimeout
from spotipy.exceptions import SpotifyException

CLIENT_ID = "4109b2d9f5014671863bb2df1d1fbdd3"
CLIENT_SECRET = "5cb719fae80c4191a9c0b288f51bfe50"

sp = spotipy.Spotify(
    auth_manager=SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET),
    requests_timeout=5
)

artists = ["Los Fabulosos Cadillacs", "Manu Chao", "Muchachito Bombo Infierno"]

def get_artist_id(name):
    try:
        result = sp.search(q=name, type='artist', limit=1)
        return result['artists']['items'][0]['id'] if result['artists']['items'] else None
    except:
        return None

def get_top_tracks(artist_name):
    try:
        artist_id = get_artist_id(artist_name)
        if not artist_id:
            return []
        tracks = sp.artist_top_tracks(artist_id, country='US')
        return [t['name'] for t in tracks['tracks'][:5]]
    except:
        return []

def get_related_artists(artist_name):
    try:
        artist_id = get_artist_id(artist_name)
        if not artist_id:
            return []
        related = sp.artist_related_artists(artist_id)
        return [a['name'] for a in related['artists'][:5]]
    except SpotifyException as e:
        print(f"Related artist error for '{artist_name}': {e.http_status}")
        return []
    except ReadTimeout:
        print(f"Timeout for '{artist_name}'")
        return []
    except:
        return []

# Top tracks from favorite artists
top_tracks = [get_top_tracks(artist) for artist in artists]

# Related artists
related_artists = [get_related_artists(artist) for artist in artists]

# Top tracks from related artists
related_top_tracks = []
for sublist in related_artists:
    for rel in sublist:
        related_top_tracks.append(get_top_tracks(rel))

# Flatten
all_top_tracks = [track for sublist in top_tracks for track in sublist]
all_related_tracks = [track for sublist in related_top_tracks for track in sublist]

playlist_tracks = all_top_tracks + all_related_tracks

# Try to fetch featured playlists
try:
    featured = sp.featured_playlists(limit=5)
    featured_playlists = [(pl['name'], pl['id']) for pl in featured['playlists']['items']]
    print("Top 5 Featured Playlists:")
    for name, pid in featured_playlists:
        print(f"{name} - {pid}")
except SpotifyException as e:
    print(f"Featured playlists error: {e.http_status}")
except:
    print("Could not load featured playlists.")

```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

## Exercise 2: Unraveling the World of Playlists


1. **Featured Exploration**: 
   - Fetch the list of Spotify's current featured playlists. 
   - Extract and display the names and IDs of the top 5 featured playlists.
   
2. **Deep Dive**:

   - Fetch and display its name, description, and total track count.

3. **Track-tastic**:

   - Create a dictionary where the keys are the names of the first 10 tracks, and the values are lists containing the names of the artists associated with each track.
   - For example: `{"TrackName1": ["Artist1", "Artist2"], "TrackName2": ["Artist3"]}`
   

```python
# Your answer here
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from spotipy.exceptions import SpotifyException
import time
import random

# Auth setup
client_id = 'your_client_id'
client_secret = 'your_client_secret'

auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(auth_manager=auth_manager)

# Favorite artists
artists = [
    "Manu Chao",
    "Los Fabulosos Cadillacs",
    "Muchachito Bombo Infierno"
]

# Get artist ID by name
def get_artist_id(name):
    result = sp.search(q='artist:' + name, type='artist', limit=1)
    items = result['artists']['items']
    return items[0]['id'] if items else None

# Get top 3 tracks for the artist
def get_top_tracks(artist_id):
    try:
        tracks = sp.artist_top_tracks(artist_id)
        return [t['name'] for t in tracks['tracks'][:3]]
    except SpotifyException as e:
        print(f"Top tracks error for {artist_id}: {e}")
        return []

# Get related artists and handle potential 404 errors
def get_related_artists(artist_id):
    try:
        related = sp.artist_related_artists(artist_id)
        return [a['name'] for a in related['artists'][:3]]
    except SpotifyException as e:
        print(f"Related artist error for {artist_id}: {e}")
        return []
    except Exception as e:
        print(f"Unexpected error for {artist_id}: {e}")
        return []

# Collect top tracks from favorite artists
favorite_tracks = []
for name in artists:
    artist_id = get_artist_id(name)
    if artist_id:
        tracks = get_top_tracks(artist_id)
        favorite_tracks.extend(tracks)

# Collect related artists and their top tracks
related_tracks = []
for name in artists:
    artist_id = get_artist_id(name)
    if artist_id:
        related = get_related_artists(artist_id)
        for rel_name in related:
            rel_id = get_artist_id(rel_name)
            if rel_id:
                tracks = get_top_tracks(rel_id)
                related_tracks.extend(tracks)
            time.sleep(random.uniform(1, 3))  # Random delay to avoid rate limiting

# Combine favorite and related tracks
all_tracks = list(set(favorite_tracks + related_tracks))

# Print collected tracks
print("Collected Tracks:")
for t in all_tracks:
    print("-", t)

# Fetch featured playlists with increased timeout
try:
    featured = sp.featured_playlists(country='US', locale='en_US', limit=5)
    print("\nTop 5 Featured Playlists:")
    for pl in featured['playlists']['items']:
        print(f"- {pl['name']} ({pl['id']})")
except SpotifyException as e:
    print(f"Featured playlists error: {e}")
except Exception as e:
    print(f"Unexpected error fetching playlists: {e}")




```
